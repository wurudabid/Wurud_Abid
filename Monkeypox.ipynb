## Exploratory Data Analysis

# Top 10 Countries based on Confirmed Cases

```python
#sorting the dataset  
data_sorted = data.sort_values(by=['Confirmed_Cases'])

#getting the first 10 values 
d1 = data_sorted[104:] 

#plot the cases 
fig = px.bar(x='Country', y='Confirmed_Cases', data_frame=d1, labels=['Country','Confirmed_Cases'],
             color='Confirmed_Cases',color_continuous_scale='electric',
             text_auto=True, title=f'Seven Countries with the most Confirmed Cases')
fig.show()
```

here I sort the confirmed cases values, and then I draw the plot.

# five countries with suspected case

```python
#sort suspected cases 
sus = data.sort_values(by=['Suspected_Cases']).iloc[-5:]

#divide lebes from features  
labels = np.array(sus['Country'])
max = np.array(sus['Suspected_Cases'])

#plot
plt.xkcd()
area = (3 * 7) ** 2
colors = np.random.rand(5)
fig, ax = plt.subplots()
ax.plot(max, labels, 'r--')
ax.scatter(max, labels, s=area, alpha=0.5, c=colors)
plt.ylabel('Suspected_Cases ')
plt.xlabel('Countries')
plt.title('5 countries with suspected cases ')
plt.show()
```

In this plot we can see the most suspected cases of monkeypox, I use scatter plot to show how the cases going up.

#Travel history of the United States
This pie plot represents the percentage of peoples in the United States that traveled with and without the monkeypox.

```python
us = data.loc[3]
cases =[us['Travel_History_Yes'], us['Travel_History_No']]
name = ['Travel_History_Yes','Travel_History_No']
fig = px.pie(us, values= cases, names=name, title='travel history of the United States',color_discrete_sequence=px.colors.sequential.RdBu)
fig.show()
```

#Travel history of confirmed cases 
here I represented the travel history of people how travel with a monkeypox, in a pie plot because it represented the percentage of infected people in each country.

```python
travel = data.sort_values(by=['Travel_History_Yes'])
val = travel[-32:]['Travel_History_Yes']
cont = travel[-32:]['Country']
fig = px.pie(us, values= val, names=cont, title='travel history of confirmed cases ',color_discrete_sequence=px.colors.sequential.RdBu)
fig.show()
```
